```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
test_data_x = np.linspace(0,20,50)
test_data_y = test_data_x
squared_testData = test_data_x**2
cubic_testData = test_data_x**3
plt.plot(test_data_x, test_data_y,"green", linewidth = '3', linestyle = 'dotted', markersize = 7)
plt.style.use('dark_background')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Linear graph")
```

```python
plt.subplot(1,2,1)
plt.plot(test_data_x, squared_testData,"green", linewidth = '3', linestyle = 'dashed')

plt.xlabel("X axis")
plt.ylabel("$Y ^ 2$")
plt.title("Square graph")

plt.subplots_adjust(hspace = 5)
plt.subplot(1,2,2)    
# subplot number 2 on plot of 2 rows and 1 column
plt.plot(test_data_x, cubic_testData,"blue", linewidth = '3', linestyle = 'dotted', )

plt.xlabel("X axis")
plt.ylabel("Y$ ^ 3$")
plt.title("Cubic graph")
```
