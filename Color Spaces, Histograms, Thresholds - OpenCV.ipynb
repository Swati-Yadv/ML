## Thresholding
allows us to separate the foreground (i.e., the objects that we are interested in) from the background of the image

Thresholding is the binarization of an image. In general, we seek to convert a grayscale image to a binary image, where the pixels are either 0 or 255.

A simple thresholding example would be selecting a threshold value T, and then setting all pixel intensities < T , to 0 and all pixel values > T, to 255. In this way, we are able to create a binary representation of the image.

Simple thresholding:  where we manually supply parameters to segment the image — this works extremely well in controlled lighting conditions where we can ensure high contrast between the foreground and background of the image.

Otsu’s thresholding:   attempt to be more dynamic and automatically compute the optimal threshold value based on the input image.

Adaptive thresholding:  instead of trying to threshold an image globally using a single value, instead breaks the image down into smaller pieces, and thresholds each of these pieces separately and individually.

```python
path1 = "C:\\Users\\Swati\\Python_Basics\\sample_images\\lena.jpg"
img = cv2.imread(path1,0)
cv2.imshow("Original image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# cv2.imread(path, flag)
# flag :    It specifies the way in which image should be read. It’s default value is cv2.IMREAD_COLOR
# cv2.IMREAD_COLOR:   load a color image, pass integer 1 for it.
# cv2.IMREAD_GRAYSCALE: It specifies to load an image in grayscale mode. we can pass integer value 0 for this flag.
# cv2.IMREAD_UNCHANGED: It specifies to load an image as such including alpha channel.we can pass integer value -1 for this flag.
```

This method requires 4 arguments.

1st arg:  the grayscale image that we wish to threshold.

2nd arg:  manually supply our T threshold value.  T=110.

3rd arg:  the output value applied during thresholding. 
any pixel value > 110 is set to 0. Any value < 110 is set to 255.

4th arg: provide a thresholding method.

```python
ret, thresh1 = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY_INV)
cv2.imshow("Thresholded image", thresh1)
cv2.imshow("Inverse Thresholded image", thresh2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Supported parameters for threshold function
cv2.THRESH_BINARY
cv2.THRESH_BINARY_INV
cv2.THRESH_TRUNC
cv2.THRESH_TOZERO
cv2.THRESH_TOZERO_INV

## Ostu Binarization

 A bimodal range is an image whose histogram has two peaks
We can approximately take a value in the middle of those peaks as threshold value
Ostu binarization does automatically calculates the optimal threshold value from the image

```python
path1 = "C:\\Users\\Swati\\Python_Basics\\sample_images\\apple.jpg"
img = cv2.imread(path1,0)
cv2.imshow("Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, thresh1 = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY)  #simple thresholding
ret, thresh2 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)  #otsu binarization
print(ret)
cv2.imshow("Threshold image", thresh1)
cv2.imshow("Ostu threshold image", thresh2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Adaptive Thresholding

Simple thresholding may not be good in all the conditions if image has different lightning conditions in different areas.
Adaptive thresholding calculates the threshold for small regions of the image.
So different thresholds for different regions of the same image are calculated which gives better results for images with varying illumination.

```python
path1 = "C:\\Users\\Swati\\Python_Basics\\sample_images\\sudoku.png"
img = cv2.imread(path1, 0)
cv2.imshow("Original img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
cv2.imshow("Global threshold", th1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 4)
#                           image, th value, adaptive method, threshold method, block size, C
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 4)
cv2.imshow("Adaptive mean threshold", th2)
cv2.imshow("Adaptive Gaussian threshold", th3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Adaptive method - Decides how thresholding value is calculated
cv2.ADAPTIVE_THRESH_MEAN_C :   threshold value is the mean of neighbourhood areas
cv2.ADAPTIVE_THRESH_GAUSSIAN_C: threshold value is the weighted sum of neighbourhood values where weights are a gaussian window

Block Size: It decides the size of neighbourhood area
C - It is a constant which is subtracted from the mean or weighted mean calculated
